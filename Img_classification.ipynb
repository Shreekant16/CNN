```python
import numpy as np
import pandas as pd
import os
from skimage.io import imread
from skimage.transform import resize
```

```python
data = 'C:\Images'
```

```python
category = ['dog','fish','butterfly']
flat_data = []
target = []
images = []
for i in category:
    paths = os.path.join(data,i)
    target_value = category.index(i)
    #print(paths)
    for img in os.listdir(paths):
        img_paths = os.path.join(paths,img)
        img = imread(img_paths)
        img_resize = resize(img,(150,150,3))
        images.append(img_resize)
        flat_data.append(img_resize.flatten())
        target.append(target_value)
target =  np.array(target)
images = np.array(images)
flat_data = np.array(flat_data)
```

```python
dataset = pd.DataFrame(flat_data)
dataset['target'] = target
dataset
```

```python
x = dataset.iloc[:,:67500].values
y = dataset.iloc[:,67500].values
print(x)
# print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=5)
```

```python
from sklearn.svm import SVC
classifier = SVC()
from sklearn.model_selection import GridSearchCV
parameters = [{'C':[1,10,100,1000],'kernel':['linear'],
               'C':[1,10,100,1000],'kernel':['rbf'],'gamma':[0.0,0.1,0.01,0.001,1,2],
               'C':[1,10,100,1000],'kernel':['rbf'],'gamma':['scale']
              }]
search = GridSearchCV(estimator=classifier,param_grid=parameters,scoring='accuracy',cv=10)
search = search.fit(x_train, y_train)
```

```python
search.best_score_
```

```python
search.best_params_
```

```python
from sklearn.svm import SVC
clss = SVC(C=1,gamma='scale',kernel='rbf')
clss.fit(x_train, y_train)
```

```python
y_pred = clss.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(f"Accuracy score {accuracy_score(y_test, y_pred)*100}")
import seaborn as sns
sns.heatmap(cm,annot=True)
```
